# Pairwise geometry-generating operations

```python
import shapely
import geopandas as gpd
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = shapely.Point((0,0)).buffer(1)
y = shapely.Point((1,0)).buffer(1)
shapely.GeometryCollection([x,y])
```

```python
x.intersection(y)
```

```python
x.difference(y)
```

```python
x.union(y)
```

```python
x.symmetric_difference(y)
```

```python
geom1 = gpd.GeoSeries(x)
geom2 = geom1.translate(0, 1)
geom3 = geom1.translate(0,2)
geom = pd.concat([geom1, geom2, geom3])
geom
```

```python
ax = geom.plot(color='#00000030', ec='k')
gpd.GeoSeries(y).plot(ax=ax, color='#FF000040', ec='k');
```

```python
geom_inter_y = geom.intersection(y)
geom_inter_y
```

```python
geom_inter_y.plot(color='y', ec='k', alpha=0.3);
```
